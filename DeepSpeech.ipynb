```
#make csvs
import os
f1=open('train/train.csv','w')
f2=open('validate/validate.csv','w')
f3=open('test/test.csv','w')
f=open('full.csv','r')
lines=f.readlines()
f1.write(lines[0])
f2.write(lines[0])
f3.write(lines[0])
for line in lines[1:]:
  fname=line.split(',')[0]
  if(fname in os.listdir('train')):
    f1.write(line)
  elif(fname in os.listdir('validate')):
    f2.write(line)
  if(fname in os.listdir('validate')):
    f3.write(line)
f1.close()
f2.close()
f3.close()
f.close()
```

```
#make vocab
f=open('full.csv','r')
f1=open('vocabulary.txt','w')
lines=f.readlines()[1:]
for line in lines:
  text=line.split(',')[2]
  f1.write(text)
f.close()
f1.close()
```

```
!unzip /content/drive/My\ Drive/SRU_Final/test.zip
!unzip /content/drive/My\ Drive/SRU_Final/train.zip
!unzip /content/drive/My\ Drive/SRU_Final/validate.zip
# !cp /content/drive/My\ Drive/SRU_Final/test.csv test2/
# !cp /content/drive/My\ Drive/SRU_Final/train.csv train2/
# !cp /content/drive/My\ Drive/SRU_Final/validate.csv validate2/
# #uncomment everything in this block when running for the first time

!sudo apt-get install sox
import os
import sox
import librosa
import soundfile as sf
from pathlib import Path
tfm = sox.Transformer()
tfm.channels(1)
tfm.rate(16000)
!mkdir train2
!mkdir test2
!mkdir validate2
!cp /content/drive/My\ Drive/SRU_Final/test.csv test2/
!cp /content/drive/My\ Drive/SRU_Final/train.csv train2/
!cp /content/drive/My\ Drive/SRU_Final/validate.csv validate2/
i=0
for wav in os.listdir('train/'):
  if(wav[-3:]=='wav'):
    print(i)
    i+=1
    x,_ = librosa.load('train/'+wav,sr=16000)
    sf.write('train/'+wav, x, 16000)
    tfm.build('train/'+wav,'train2/'+wav)
i=0
for wav in os.listdir('test/'):
  if(wav[-3:]=='wav'):
    print(i)
    i+=1
    x,_ = librosa.load('test/'+wav,sr=16000)
    sf.write('test/'+wav, x, 16000)
    tfm.build('test/'+wav,'test2/'+wav)
i=0
for wav in os.listdir('validate/'):
  if(wav[-3:]=='wav'):
    print(i)
    i+=1
    x,_ = librosa.load('validate/'+wav,sr=16000)
    sf.write('validate/'+wav, x, 16000)
    tfm.build('validate/'+wav,'validate2/'+wav)
```
